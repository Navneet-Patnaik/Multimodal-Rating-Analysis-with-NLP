```python
# ==================== 1. Imports ====================
import pandas as pd
import numpy as np
import string
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sentence_transformers import SentenceTransformer

# ==================== 2. Load and Explore Data ====================
df = pd.read_excel("Artificial_Data.xlsx")
print(df.head(5))

print("Shape:", df.shape)
print(df.dtypes.value_counts())
print(df['Rating'].value_counts())
print(df['RATING_TYPE'].value_counts())

# ==================== 3. Preprocessing ====================
# Label encode Rating
rating_encoder = LabelEncoder()
df['Rating_encoded'] = rating_encoder.fit_transform(df['Rating'])

# One-hot encode RATING_TYPE
df = pd.get_dummies(df, columns=['RATING_TYPE'], drop_first=True)

# Scale structured features
structured_features = df.drop(columns=['Rating', 'Rating_encoded', 'string_values'])
scaler = StandardScaler()
scaled = scaler.fit_transform(structured_features)
cleaned_df = pd.DataFrame(scaled, columns=structured_features.columns)
cleaned_df['Rating_encoded'] = df['Rating_encoded']

# ==================== 4. Text Preprocessing ====================
def clean_text(text):
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    return " ".join([word for word in text.split() if word not in ENGLISH_STOP_WORDS])

df['cleaned_text'] = df['string_values'].apply(clean_text)

# ==================== 5. TF-IDF Embeddings ====================
tfidf_vectorizer = TfidfVectorizer(max_features=100)
tfidf_matrix = tfidf_vectorizer.fit_transform(df['cleaned_text'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# ==================== 6. Sentiment Analysis ====================
df['sentiment_score'] = df['cleaned_text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# ==================== 7. Sentence-BERT Embeddings ====================
sbert = SentenceTransformer('all-MiniLM-L6-v2')
bert_embeddings = sbert.encode(df['cleaned_text'].tolist(), show_progress_bar=True)
bert_df = pd.DataFrame(bert_embeddings)

# ==================== 8. Model Prep ====================
y = df['Rating_encoded']
X_struct = cleaned_df.drop(columns='Rating_encoded')
X_text_tfidf = tfidf_df
X_text_bert = bert_df
X_combined = pd.concat([X_struct.reset_index(drop=True), tfidf_df], axis=1)
X_combined['sentiment_score'] = df['sentiment_score']
X_combined_bert = pd.concat([X_struct.reset_index(drop=True), bert_df], axis=1)
X_combined_bert.columns = X_combined_bert.columns.astype(str)

# ==================== 9. Train-Test Split ====================
X_train_s, X_test_s, y_train, y_test = train_test_split(X_struct, y, test_size=0.2, random_state=42)
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_text_tfidf, y, test_size=0.2, random_state=42)
X_train_comb, X_test_comb, _, _ = train_test_split(X_combined, y, test_size=0.2, random_state=42)
X_train_bert, X_test_bert, _, _ = train_test_split(X_text_bert, y, test_size=0.2, random_state=42)
X_train_comb_bert, X_test_comb_bert, _, _ = train_test_split(X_combined_bert, y, test_size=0.2, random_state=42)

# ==================== 10. Train & Evaluate Models ====================
def evaluate_model(name, model, X_train, X_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{name} Report:")
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{name} Confusion Matrix')
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# Structured-only
evaluate_model("Structured (RandomForest)", RandomForestClassifier(random_state=42), X_train_s, X_test_s)

# Text-only (TF-IDF)
evaluate_model("Text TF-IDF (LogisticRegression)", LogisticRegression(max_iter=1000), X_train_tfidf, X_test_tfidf)

# Combined (TF-IDF + structured + sentiment)
evaluate_model("Combined TF-IDF+Sentiment (RandomForest)", RandomForestClassifier(random_state=42), X_train_comb, X_test_comb)

# Text-only (BERT)
evaluate_model("Text BERT (LogisticRegression)", LogisticRegression(max_iter=1000), X_train_bert, X_test_bert)

# Combined (BERT + structured)
evaluate_model("Combined BERT (RandomForest)", RandomForestClassifier(random_state=42), X_train_comb_bert, X_test_comb_bert)
```

```python
import numpy as np
import pandas as pd
import string
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# --- Load and Clean Text ---
def clean_text(text):
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    return " ".join([word for word in text.split() if word not in ENGLISH_STOP_WORDS])

df = pd.read_excel("Artificial_Data.xlsx")
df['cleaned_text'] = df['string_values'].apply(clean_text)

# --- Load GloVe Vectors ---
def load_glove_embeddings(glove_path="glove.6B.100d.txt"):
    embeddings = {}
    with open(glove_path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            embeddings[word] = vector
    return embeddings

glove_embeddings = load_glove_embeddings("glove.6B.100d.txt")

# --- Create Sentence Embeddings ---
def sentence_embedding(text, embeddings, dim=100):
    words = text.split()
    valid_vectors = [embeddings[word] for word in words if word in embeddings]
    if valid_vectors:
        return np.mean(valid_vectors, axis=0)
    else:
        return np.zeros(dim)

glove_vectors = np.vstack(df['cleaned_text'].apply(lambda x: sentence_embedding(x, glove_embeddings)))

# Convert to DataFrame
glove_df = pd.DataFrame(glove_vectors)
glove_df.head()
```

```python
import pandas as pd
import numpy as np
import string
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# === 1. Load & Preprocess ===
df = pd.read_excel("Artificial_Data.xlsx")

def clean_text(text):
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    return " ".join([word for word in text.split() if word not in set(tf.keras.utils.get_file('stopwords', origin='https://raw.githubusercontent.com/stopwords-iso/stopwords-en/master/stopwords-en.txt').splitlines())])

df['cleaned_text'] = df['string_values'].apply(clean_text)

# === 2. Encode Target ===
label_encoder = LabelEncoder()
df['Rating_encoded'] = label_encoder.fit_transform(df['Rating'])

# === 3. Tokenize & Pad ===
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(df['cleaned_text'])
sequences = tokenizer.texts_to_sequences(df['cleaned_text'])
word_index = tokenizer.word_index
padded_sequences = pad_sequences(sequences, padding='post', maxlen=50)

# === 4. Load GloVe 100d Embeddings ===
embedding_index = {}
with open("glove.6B.100d.txt", 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs

# === 5. Build Embedding Matrix ===
embedding_dim = 100
vocab_size = len(word_index) + 1
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embedding_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

# === 6. Train-Test Split ===
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, df['Rating_encoded'], test_size=0.2, random_state=42)

# === 7. BiLSTM Model ===
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, weights=[embedding_matrix], input_length=50, trainable=False),
    Bidirectional(LSTM(64, return_sequences=False)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(len(label_encoder.classes_), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# === 8. Train ===
history = model.fit(X_train, y_train, validation_split=0.2, epochs=5, batch_size=32)

# === 9. Evaluate ===
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from cleantext import clean
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the dataset
file_path = "Artificial_Data.xlsx"  # Update this path if needed
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name="Sheet1")


# Step 1: Transform rating classes
dict_trans_class = {
    1: [1, 2, 3, 4, 5],
    10: [10, 11, 12],
    13: [13, 14, 15],
    16: [16, 17, 18, 19, 20, 21, 22]
}

lbl_first = df["Rating"].values
lbl_final = lbl_first.copy()

for main_class, sub_classes in dict_trans_class.items():
    for sub_class in sub_classes:
        lbl_final = np.where(lbl_first == sub_class, main_class, lbl_final)

# Convert remaining unique values into sequential indices
lbl_final_rest = lbl_final.copy()
unique_classes = np.unique(lbl_final)
for idx, class_value in enumerate(unique_classes):
    lbl_final_rest = np.where(lbl_final == class_value, idx, lbl_final_rest)

# Convert labels to categorical format using NumPy (instead of TensorFlow)
num_classes = len(np.unique(lbl_final_rest))
lbl_binary = np.eye(num_classes)[lbl_final_rest.astype(int)]

# Step 2: Text Preprocessing Functions
def preprocess_txt(text):
    REPLACE_NO_SPACE = re.compile(r"(\.|;|:|!|\?|,|\"|\(|\)|\[|\]|\d+)")
    REPLACE_WITH_SPACE = re.compile(r"(<br\s*/><br\s*/>)|(\-)|(\/)")
    text = REPLACE_NO_SPACE.sub("", str(text).lower())
    text = REPLACE_WITH_SPACE.sub(" ", text)
    return text

def clean_text(sent):
    return clean(sent, lower=True, no_urls=True, no_emails=True, no_phone_numbers=True,
                 no_digits=True, no_currency_symbols=True, no_punct=True, lang="en")

def remove_stop_words(text):
    stop_words = set(stopwords.words('english'))
    return ' '.join([word for word in text.split() if word not in stop_words])

def get_stemmed_text(text):
    stemmer = PorterStemmer()
    return ' '.join([stemmer.stem(word) for word in text.split()])

# Apply Text Preprocessing
df["string_values"] = df["string_values"].astype(str).apply(clean_text)
df["string_values"] = df["string_values"].apply(preprocess_txt)
df["string_values"] = df["string_values"].apply(remove_stop_words)
df["string_values"] = df["string_values"].apply(get_stemmed_text)

# Step 3: Tokenization & Padding
MAX_NB_WORDS = 500000
tokenizer = Tokenizer(num_words=MAX_NB_WORDS)
tokenizer.fit_on_texts(df["string_values"])
data_seq = tokenizer.texts_to_sequences(df["string_values"])
vocab_size = len(tokenizer.word_index) + 1

maxlen = 5000  # Set maximum sequence length
data_seq_pad = pad_sequences(data_seq, padding='pre', maxlen=maxlen)

# Step 4: Numeric Feature Processing
# Select only numerical columns
num_features = df.select_dtypes(include=['number'])


scaler = MinMaxScaler()
num_features_scaled = pd.DataFrame(scaler.fit_transform(num_features), columns=num_features.columns)

# Step 5: Combine Metadata
meta_data = pd.concat([num_features_scaled], axis=1).values

# Step 6: Train-Test Split
train_percentage = 0.80
x_train, x_test, y_train, y_test = train_test_split(data_seq_pad, lbl_binary, test_size=1-train_percentage, random_state=0)
x_train_meta, x_test_meta, _, _ = train_test_split(meta_data, lbl_binary, test_size=1-train_percentage, random_state=0)

# Reshape Metadata
x_train_meta = x_train_meta.reshape(x_train_meta.shape[0], x_train_meta.shape[1], 1)
x_test_meta = x_test_meta.reshape(x_test_meta.shape[0], x_test_meta.shape[1], 1)

# Step 7: Class Weights Calculation
class_weights = class_weight.compute_class_weight(class_weight="balanced",
                                                  classes=np.unique(np.argmax(y_test, axis=1)),
                                                  y=np.argmax(y_test, axis=1))
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Output Results
print("-" * 50)
print(f"Train shape: {x_train.shape}")
print(f"Test shape: {x_test.shape}")
print(f"x_train_meta shape: {x_train_meta.shape}")
print(f"x_test_meta shape: {x_test_meta.shape}")
print("Class Weights:", class_weight_dict)
print("-" * 50)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Dropout, Bidirectional, Flatten, Concatenate

# Model Hyperparameters
embedding_dim = 500
lstm_units = 128
dense_units = 64
dropout_rate = 0.3
n_classes = lbl_binary.shape[1]

# Define Text Input Branch
text_input = Input(shape=(maxlen,), name="text_input")
embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=maxlen)(text_input)
lstm = Bidirectional(LSTM(lstm_units, return_sequences=False))(embedding)
text_branch = Dense(128, activation='relu')(lstm)

# Define Metadata Input Branch
meta_input = Input(shape=(x_train_meta.shape[1], 1), name="meta_input")
meta_branch = Flatten()(meta_input)
meta_branch = Dense(128, activation='relu')(meta_branch)

# Merge Both Branches
merged = Concatenate()([text_branch, meta_branch])
merged = Dense(dense_units, activation='relu')(merged)
merged = Dropout(dropout_rate)(merged)
output = Dense(n_classes, activation='softmax', name="output_layer")(merged)

# Build & Compile Model
model = Model(inputs=[text_input, meta_input], outputs=output)
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy", keras.metrics.AUC(name="auc")])

# Model Summary
model.summary()

# Train the Model
history = model.fit(
    [x_train, x_train_meta], y_train,
    validation_data=([x_test, x_test_meta], y_test),
    batch_size=32,
    epochs=10,
    class_weight=class_weight_dict
)

# Model Evaluation
results = model.evaluate([x_test, x_test_meta], y_test)
print("Test Loss:", results[0])
print("Test Accuracy:", results[1])
print("Test AUC:", results[2])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor

# Load main + enriched dataset
df = pd.read_csv("data_with_dominant_topic.csv")
df["Rating_encoded"] = LabelEncoder().fit_transform(df["Rating"])

# ---- TEXT ----
tfidf = TfidfVectorizer(max_features=300)
text_features = tfidf.fit_transform(df["string_values"].astype(str)).toarray()

# Reduce text dimensionality
pca = PCA(n_components=50, random_state=42)
text_features_pca = pca.fit_transform(text_features)

# ---- STRUCTURED ----
exclude = ["Rating", "string_values", "Rating_encoded", "Grouped_Class", "sentiment_label"]
X_structured = df.drop(columns=exclude, errors='ignore')

X_structured = X_structured.select_dtypes(include=['number'])

# Encode topic as one-hot
X_topic = pd.get_dummies(df["dominant_topic"], prefix="topic")

# Normalize structured features
scaler = MinMaxScaler()
X_structured_scaled = scaler.fit_transform(X_structured)

# ---- FINAL INPUT SET ----
X_combined = np.concatenate([X_structured_scaled, text_features_pca, X_topic.values], axis=1)
y = df["Rating_encoded"]

# ---- TRAIN/TEST SPLIT ----
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

# ---- MODELS ----
def evaluate_model(name, y_true, y_pred):
    print(f"Model: {name}")
    print("R2 Score: ", round(r2_score(y_true, y_pred), 4))
    print("MAE     : ", round(mean_absolute_error(y_true, y_pred), 4))
    print("RMSE    : ", round(np.sqrt(mean_squared_error(y_true, y_pred)), 4))
    print("-" * 30)

# Random Forest
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
evaluate_model("Random Forest", y_test, rf.predict(X_test))

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
evaluate_model("Linear Regression", y_test, lr.predict(X_test))

# XGBoost
xgb = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.1)
xgb.fit(X_train, y_train)
evaluate_model("XGBoost", y_test, xgb.predict(X_test))
```
