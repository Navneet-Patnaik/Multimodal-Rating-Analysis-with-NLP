```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Artificial_Data.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Class Distribution

```python
df['Rating'].value_counts().plot(kind='bar', title='Original Rating Distribution', figsize=(8,5))
```

```python
df.describe(include='all')
```

## Correlation Heatmap

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=False, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```
